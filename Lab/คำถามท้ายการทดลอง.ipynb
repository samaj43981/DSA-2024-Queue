#คำถามท้ายการทดลอง
##1. Queue ต่างจาก Stack อย่างไร?
     Queue เป็นการใส่ข้อมูลแบบ FIFO (First In First Out)
     Stack: เป็นการใส่ข้อมูลแบบ LIFO (Last In First Out)


##2. เพราะเหตุใดการ dequeue จึงใช้ pop(0) แทนที่จะเป็น pop()?
     เพราะ pop(0) นำข้อมูลจากหน้าคิวออก ตรงตาม Queue ที่เป็นหลักการFIFO


##3. ยกตัวอย่างการประยุกต์ใช้ Queue ในชีวิตประจำวันมา 3 ตัวอย่าง
    การรอคิวซื้อของ
    สั่งปริ้นเอกสารในเครื่องปริ้น
    ระบบการต่อแถวซื้อบัตรคอน


##4. หากต้องการทำ Priority Queue จะต้องปรับปรุงโค้ดอย่างไร?
    เพิ่มการจัดการ ลำดับความสำคัญ (Priority)โดยใช้ ลิสต์ (List) หรือ heapq ใน Python เพื่อจัดการข้อมูลตามลำดับความสำคัญ
